```python
import pandas as pd
import numpy as np
import sklearn
```

```python
df = pd.read_csv("/content/ECG200_TRAIN.csv",header = None, delimiter = '  ' )
```

```python
df
```

```python
mone = df[df[0] == -1]
mone = mone.iloc[:2,:]
df = df.append(mone)
```

```python
one = df[df[0] == 1]
mone = df[df[0] != 1]
```

```python
one
```

# Numpy Splitting

```python
spl_one = np.array_split(one,3)
spl_mone = np.array_split(mone,3)
```

```python
spl_mone
```

## **Fold 1**

```python
train = pd.DataFrame()
train = train.append(spl_mone[0:2])
train = train.append(spl_one[0:2])

train
```

```python
test = pd.DataFrame()
test = test.append(spl_mone[2:])
test = test.append(spl_one[2:])

test
```

```python
trainlabel = train[0]
testlabel = test[0]
```

```python
Xtrain = train.iloc[:,1:]
Ytrain = trainlabel
Xtest = test.iloc[:,1:]
Ytest = testlabel
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(Xtrain, Ytrain)
```

```python
# Testing the classifier
y_pred = neigh.predict(Xtest)
print('Predicted',y_pred)
print('Actual data',Ytest)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,classification_report
accuracy = accuracy_score(Ytest, y_pred)*100
print('Accuracy: {0:.4f}'.format(accuracy))
print('Recall: {0:.4f}'.format(recall_score(Ytest, y_pred,average = 'macro')*100))
print('precision {0:.4f}'.format(precision_score(Ytest, y_pred,average = 'macro')*100))
print('f1_score: {0:.4f}'.format(f1_score(Ytest, y_pred,average = 'macro')*100))
```

```python
name='KNN'
print('\n--------- Result of {} -------\n'.format(name))
print(classification_report(Ytest, y_pred, digits=4))
```

# Scikit-learn

```python
df.reset_index(drop=True, inplace=True)
```

```python
X = df.iloc[:,1:]
Y = df.iloc[:,:1]
```

```python
from sklearn.model_selection import train_test_split
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.33, random_state=3)
```

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
neigh.fit(Xtrain, Ytrain)

y_pred = neigh.predict(Xtest)
```

```python
from sklearn.metrics import accuracy_score,recall_score,precision_score,f1_score,classification_report
accuracy = accuracy_score(Ytest, y_pred)*100
print('Accuracy: {0:.4f}'.format(accuracy))
print('Recall: {0:.4f}'.format(recall_score(Ytest, y_pred,average = 'macro')*100))
print('precision {0:.4f}'.format(precision_score(Ytest, y_pred,average = 'macro')*100))
print('f1_score: {0:.4f}'.format(f1_score(Ytest, y_pred,average = 'macro')*100))
```

```python
name='KNN'
print('\n--------- Result of {} -------\n'.format(name))
print(classification_report(Ytest, y_pred, digits=4))
```
