```python
import pandas as pd
```

```python
df = pd.read_csv("drinks.csv")
df
```

```python
df = df.dropna(axis = 0)
```

```python
def column(name):
    return df[name]
```

```python
data = column('beer_servings')
data
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Country = df.country
Beer_servings = data
```

```python
fig = plt.figure()
plt.hist(Beer_servings,rwidth=0.5)
plt.show()
```

```python
x = Beer_servings
y = df.total_litres_of_pure_alcohol

colors = np.random.randint(50, size=(170))
sizes = 5 * np.random.randint(50, size=(170))
plt.scatter(x, y, c=colors, s = sizes ,alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show() 
```

```python
country = []
total = []

total_al = list(df.total_litres_of_pure_alcohol)
countries = list(df.country)

for i in range(0,len(countries)):
    if total_al[i] > 10:
        country.append(countries[i])
        total.append(total_al[i])
```

```python
fig = plt.figure(figsize = (33,10))
plt.bar(country,total,width = 0.7)

plt.show()
```

```python
country = []
beer = []

beer_serving = list(df.beer_servings)
countries = list(df.country)

for i in range(0,len(countries)):
    if beer_serving[i] > 100:
        country.append(countries[i])
        beer.append(beer_serving[i])
```

```python
fig = plt.figure(figsize = (15,15))
plt.pie(beer,labels=country)
plt.show()
```

```python
continent = []
total = []
spirit = []
wine = []
beer = []

total_alcohol = list(df.total_litres_of_pure_alcohol)
spirit_serving = list(df.spirit_servings)
countries = list(df.country)
wine_serving = list(df.wine_servings)
continents = list(df.continent.unique())

for i in range(0,len(continents)):
    sums = df.loc[df.continent == continents[i] , 'total_litres_of_pure_alcohol'].sum()
    if sums > 50:
        continent.append(continents[i])
        spirit.append(spirit_serving[i])
        beer.append(beer_serving[i])
        wine.append(wine_serving[i])
        total.append(total_alcohol[i])
```

```python
continent
```

```python
uni_continent = []

for i in continent:
    if i not in uni_continent:
        uni_continent.append(i)
```

```python
fig = plt.figure()

X_axis = np.arange(len(uni_continent))

plt.bar(X_axis + 0.00,spirit,width = 0.2, color = 'b')
plt.bar(X_axis + 0.25,beer,width = 0.2, color = 'black')
plt.bar(X_axis + 0.50,wine,width = 0.2, color = 'g')

plt.show()
```
